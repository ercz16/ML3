# Step 1: Split the dataset into training and testing sets

```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical


# Set the path to the dataset folder
data_folder = './data_folder/'

# Collect audio files
audio_files = []
for folder in os.listdir(data_folder):
    if folder[0] == ".":
        continue
    fp = os.path.join(data_folder, folder)
    audio_files += [os.path.join(fp, f) for f in os.listdir(fp) if f.endswith('.wav')]


# Define the list of emotion labels
labels = ["sad", "angry", "happy", "fear"]
    
# Create a list of emotion labels for each audio file
emotion_labels = []
for file in audio_files:
    for emotion in labels:
        if emotion in file:
            emotion_labels.append(emotion)
            break

            
# Encode the categorical labels as integers
label_encoder = LabelEncoder()
emotion_encoded = label_encoder.fit_transform(emotion_labels)

# Convert the integer labels to one-hot encoded vectors
emotion_onehot = to_categorical(emotion_encoded)
```

```python
# Convert to numpy
audio_np = np.array(audio_files)
```

```python
# Split the dataset into training and testing sets
X_train_files, X_test_files, y_train, y_test = train_test_split(audio_np, emotion_onehot, test_size=0.3, random_state=42, stratify=emotion_labels)
```

# Step 2: Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the label distribution of the dataset
plt.figure(figsize=(8,6))
sns.countplot(y_train)
plt.title('Label Distribution of Training Dataset')
plt.xlabel('Number of Audio Clips')
plt.ylabel('Emotion')
plt.show()
```

# Step 3: Acoustic Feature Extraction

```python
import librosa.display
import numpy as np

# Load a sample audio clip
audio_file = X_train_files[0]
y, sr = librosa.load(audio_file)

# Plot the waveform of the audio clip
plt.figure(figsize=(8,6))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()

# Plot the spectrogram of the audio clip
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
log_S = librosa.power_to_db(S, ref=np.max)
plt.figure(figsize=(8,6))
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')
plt.title('Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```

# Step 4: Feature Post-processing

```python
# Define a function to extract acoustic features from an audio clip
def extract_features(audio_file):
    # Load the audio clip
    y, sr = librosa.load(audio_file)
    # Define window parameters
    frame_length = 0.025  # 25 milliseconds
    frame_step = 0.01  # 10 milliseconds

    # Compute MFCCs with time windows
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13, n_fft=int(frame_length * sr),
                              hop_length=int(frame_step * sr), window='hamming')
    return mfccs

def resize_features(features, max_len):
    resized = np.zeros((len(features), max_len))
    for i, f in enumerate(features):
        resized[i][:len(f)] = f
    return resized
```

```python
X_train = [extract_features(f) for f in X_train_files]
X_test = [extract_features(f) for f in X_test_files]

train_max = max([f.shape for f in X_train])[1]
test_max = max([f.shape for f in X_test])[1]
```

```python
for i in range(len(X_train)):
    X_train[i] = resize_features(X_train[i], train_max)
    
for i in range(len(X_test)):
    X_test[i] = resize_features(X_test[i], test_max)
    
```

```python
train_X = np.array(X_train)
test_X = np.array(X_test)

train_X = train_X.reshape(train_X.shape[0], -1)
test_X = test_X.reshape(test_X.shape[0], -1)
```

```python
train_y = []
for l in y_train:
    for i, num in enumerate(l):
        if num == 1.0:
            train_y.append(i+1)
            break

test_y = []
for l in y_test:
    for i, num in enumerate(l):
        if num == 1.0:
            test_y.append(i+1)
            break
```

```python
test_y = np.array(test_y)
```

# Step 5: Build your audio emotion recognition model

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# Train a SVM classifier on the training data
clf = SVC(kernel='linear')
clf.fit(train_X, train_y)

# Evaluate the performance of the model on the testing data
y_pred = clf.predict(test_y.reshape(-1,1))
# print(classification_report(y_test, y_pred))
# print(confusion_matrix(y_test, y_pred))
```
